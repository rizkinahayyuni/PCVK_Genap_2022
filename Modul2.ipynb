1. Tampilkan image dalam channel Red-Green dan Green-Blue saja!

```python
image = cv.imread('/content/drive/MyDrive/DATA/kitten02.jpg')
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image2)
```

```python
red_green = image2.copy() # Duplicate image
red_green[:, :, 2] = 0    # Zero out contribution from blue
plt.imshow(red_green)
```

```python
green_blue = image2.copy()  # Duplicate image
green_blue[:, :, 0] = 0     # Zero out contribution from red
plt.imshow(green_blue)
```

2. Tampilkan image baris ke 10-100, kolom 10-100!

4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

```python
import random

def ranInt():
  return random.randint(0, 512)

black_img3 = np.zeros(shape=(512,512,3), dtype=np.int32)

cv.rectangle(black_img3, pt1=(ranInt(),ranInt()),pt2=(ranInt(),ranInt()),color=(255,0,0),thickness=10)
cv.rectangle(black_img3, pt1=(ranInt(),ranInt()),pt2=(ranInt(),ranInt()),color=(0,255,0),thickness=10)
cv.rectangle(black_img3, pt1=(ranInt(),ranInt()),pt2=(ranInt(),ranInt()),color=(0,0,255),thickness=10)
cv.rectangle(black_img3, pt1=(ranInt(),ranInt()),pt2=(ranInt(),ranInt()),color=(255,0,255),thickness=10)
cv.rectangle(black_img3, pt1=(ranInt(),ranInt()),pt2=(ranInt(),ranInt()),color=(0,255,255),thickness=10)

plt.imshow(black_img3)
```
