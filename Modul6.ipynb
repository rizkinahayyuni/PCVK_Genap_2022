```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
import osgeo
import glob
```

**Histogram Citra**

```python
img_lena = cv.imread('/content/drive/MyDrive/DATA/lena.jpg')

# Menghitung pixel warna RED([2]), mask = none, histSize(histogram sizes in each dimension) = 256, range = [0-256]
hist_r = cv.calcHist([img_lena],[2],None,[256],[0,256])     
# Menghitung pixel warna GREEN([1]), mask = none, histSize(histogram sizes in each dimension) = 256, range = [0-256]
hist_g = cv.calcHist([img_lena],[1],None,[256],[0,256])     
# Menghitung pixel warna BLUE([0]), mask = none, histSize(histogram sizes in each dimension) = 256, range = [0-256]
hist_b = cv.calcHist([img_lena],[0],None,[256],[0,256])     

fig, (ax_r, ax_g, ax_b) = plt.subplots(1, 3, figsize=[15,4], sharex=True, sharey=True)  # Membuat plot 1baris & 3kolom
fig.suptitle('Histogram RGB plot')                                                      # Menambahkan title dari plot
fig.text(0.08, 0.45, 'Jumlah Kemunculan', va='center', rotation='vertical')      # atribut(x, y, text, text_properties)
fig.text(0.5, 0.03, 'Intensitas Warna', ha='center')                             # atribut(x, y, text, text_properties)
ax_r.plot(hist_r, color='red')                    # Membuat plot untuk histogram warna merah
ax_g.plot(hist_g, color='green')                  # Membuat plot untuk histogram warna merah
ax_b.plot(hist_b, color='blue')                   # Membuat plot untuk histogram warna merah
```

**Histogram Citra dengan Numpy Histogram**

```python
rgb_lena2 = img_lena          # Copy img_lena

# bins = jumlah rentang data ; range = rentang nilai yaitu 0-256
histogram_r, bin_edges_r = np.histogram(rgb_lena2[:,:,2], bins=256, range=(0, 256))   # Histogram untuk warna RED[2]
histogram_g, bin_edges_g = np.histogram(rgb_lena2[:,:,1], bins=256, range=(0, 256))   # Histogram untuk warna GREEN[1]
histogram_b, bin_edges_b = np.histogram(rgb_lena2[:,:,0], bins=256, range=(0, 256))   # Histogram untuk warna BLUE[0]

fig, (ax_r, ax_g, ax_b) = plt.subplots(1, 3, figsize=[15,4], sharex=True, sharey=True)  # Membuat plot 1baris & 3kolom
fig.suptitle('Histogram RGB plot dengan Numpy')                                                      # Menambahkan title dari plot
fig.text(0.08, 0.45, 'Jumlah Kemunculan', va='center', rotation='vertical')      # atribut(x, y, text, text_properties)
fig.text(0.5, 0.03, 'Intensitas Warna', ha='center')                             # atribut(x, y, text, text_properties)
ax_r.plot(bin_edges_r[0:-1], histogram_r, color='red')                    # Membuat plot untuk histogram warna merah
ax_g.plot(bin_edges_g[0:-1], histogram_g, color='green')                  # Membuat plot untuk histogram warna merah
ax_b.plot(bin_edges_b[0:-1], histogram_b, color='blue')                   # Membuat plot untuk histogram warna merah
```
