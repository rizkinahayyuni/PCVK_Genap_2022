```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import PIL
import io
%matplotlib inline
from IPython.display import display, Javascript, Image
from google.colab.output import eval_js
from base64 import b64decode, b64encode
```

```python
def convert_to_RGB(image):
    return cv.cvtColor(image, cv.COLOR_BGR2RGB)

def haarcascade_face(haarcascade_frontalface,image):
  image_ori = image.copy()
  roi_wajah = haarcascade_frontalface.detectMultiScale(image)
  for(x,y,w,h) in roi_wajah:
    cv.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3)
  f, axarr = plt. subplots(1,2,figsize=(15,5))  
  axarr[0].imshow(convert_to_RGB(image_ori)), axarr[0].set_title("input image") 
  axarr[1].imshow(convert_to_RGB(image)), axarr[1].set_title("result image")

haarcascade_frontalface = cv.CascadeClassifier('/content/drive/MyDrive/DATA/haarcascades/haarcascade_frontalface_alt.xml')
jokowi = cv.imread('/content/drive/MyDrive/DATA/facedet/jokowi.jpg')
kartini = cv.imread('/content/drive/MyDrive/DATA/facedet/kartini.jpg')
kucing = cv.imread('/content/drive/MyDrive/DATA/facedet/kucing.jpg')
mask = cv.imread('/content/drive/MyDrive/DATA/facedet/mask.png')
mjordan = cv.imread('/content/drive/MyDrive/DATA/facedet/mjordan.jpg')
solvayconf = cv.imread('/content/drive/MyDrive/DATA/facedet/solvayconf.jpg')

haarcascade_face(haarcascade_frontalface,jokowi)
haarcascade_face(haarcascade_frontalface,kartini)
haarcascade_face(haarcascade_frontalface,kucing)
haarcascade_face(haarcascade_frontalface,mask)
haarcascade_face(haarcascade_frontalface,mjordan)
haarcascade_face(haarcascade_frontalface,solvayconf)
```

**Deteksi Wajah Kucing**

```python
haarcascade_frontalcatface = cv.CascadeClassifier('/content/drive/MyDrive/DATA/haarcascades/haarcascade_frontalcatface.xml')
kucing = cv.imread('/content/drive/MyDrive/DATA/facedet/kucing.jpg')
kucing_ori = kucing.copy()

roi_wajah_kucing = haarcascade_frontalcatface.detectMultiScale(kucing)

for(x,y,w,h) in roi_wajah_kucing:
  cv.rectangle(kucing,(x,y),(x+w,y+h),(0,255,0),3)

f, axarr = plt. subplots(1,2,figsize=(15,5))
axarr[0].imshow(convert_to_RGB(kucing_ori)), axarr[0].set_title("input image") 
axarr[1].imshow(convert_to_RGB(kucing)), axarr[1].set_title("result image")
```

**Eyes Detection**

```python
haarcascade_eye = cv.CascadeClassifier('/content/drive/MyDrive/DATA/haarcascades/haarcascade_eye.xml')
jokowi = cv.imread('/content/drive/MyDrive/DATA/facedet/jokowi.jpg')
jokowi_gray = cv.cvtColor(jokowi, cv.COLOR_BGR2GRAY)
jokowi_ori_gray = jokowi_gray.copy()

roi_mata = haarcascade_eye.detectMultiScale(jokowi_gray)

for(x,y,w,h) in roi_mata:
  cv.rectangle(jokowi_gray,(x,y),(x+w,y+h),(255,255,255),3)

f, axarr = plt. subplots(1,2,figsize=(15,5))
axarr[0].imshow(convert_to_RGB(jokowi_ori_gray)), axarr[0].set_title("input image") 
axarr[1].imshow(convert_to_RGB(jokowi_gray)), axarr[1].set_title("result image")
```
